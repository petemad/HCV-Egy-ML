```python
import requests


url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00503/HCV-Egy-Data.zip'
r = requests.get(url, allow_redirects=True)

open('HCV-Egy-Data.zip', 'wb').write(r.content)
```

```python
import zipfile
with zipfile.ZipFile('HCV-Egy-Data.zip',"r") as zip_ref:
    zip_ref.extractall()
```

```python
import pandas as pd
data = pd.read_csv("HCV-Egy-Data.csv")
data.describe()
```
