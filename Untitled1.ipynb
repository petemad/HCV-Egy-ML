```python
y = data1['Baselinehistological.staging']
#data1 = data1.drop(columns=['Baselinehistological.staging'])
```

```python
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.4, random_state=0)
sss.get_n_splits(data1, y)
for train_index, test_index in sss.split(data1, y):
    print(len(train_index), len(test_index), len(data1), len(y))
    x_train, x_test = data1.iloc[train_index], data.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
x_train, x_test, y_train, y_test = train_test_split(data1, y, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier(max_depth=4, random_state=0)
```

```python
clf.fit(x_train, y_train)
prediction = clf.predict(x_test)
accuracy_score(y_test, prediction)
```

```python
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train, y_train)
prediction = classifier.predict(x_test)
accuracy_score(y_test, prediction)
```

```python
clf = svm.SVC()
clf.fit(x_train, y_train)
prediction = clf.predict(x_test)
accuracy_score(y_test, prediction)
```
